```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("1.01.+Simple+linear+regression.csv")
data
```

```python
data.describe()
```

```python
y = data["GPA"]
x1 = data["SAT"]
x_matrix = x1.values.reshape(-1,1)
```

```python
plt.scatter(x_matrix,y)
plt.show()
```

```python
reg = LinearRegression().fit(x_matrix,y)
# This is to get R Squared
reg.score(x_matrix,y)
```

```python
# This is to get Coef of the feature
reg.coef_
```

```python
# This is to get intercept
reg.intercept_
```

```python
yhat = 0.0017*x1 + 0.2750
plt.scatter(x1,y)
plt.plot(x1,yhat,c="Orange", lw=4)
plt.hlines(y=y.mean(),colors="Green", xmin=1700, xmax=2000, linestyles="--")
plt.show()
```

```python
pred_data =  pd.DataFrame([3000, 3200, 3500], columns=['SAT'])
predicted_y = reg.predict(pred_data)
```

```python
#pred_data_out = pd.DataFrame()
pred_data_out = pred_data
pred_data_out["Predicted GPA"] = predicted_y
```
