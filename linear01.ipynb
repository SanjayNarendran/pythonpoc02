```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("1.01.+Simple+linear+regression.csv");
data
```

```python
data.describe()
```

```python
y = data["GPA"]
y
```

```python
x1 = data['SAT']
x1
```

```python
plt.scatter(x1,y)
plt.show()
```

```python
x = sm.add_constant(x1)
result = sm.OLS(y,x).fit()
result.summary()
```

```python
yhat = 0.0017*x1 + 0.2750
plt.scatter(x1,y)
plt.plot(x1,yhat,c="Orange", lw=4)
plt.hlines(y=y.mean(),colors="Green", xmin=1700, xmax=2000, linestyles="--")
plt.show()
```
